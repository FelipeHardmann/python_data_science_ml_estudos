## Matplotlib

```python
from matplotlib import pyplot as plt
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, gridspec_kw={'height_ratios': [1,2]}, figsize=(16, 12))
from numpy.random import randint
import numpy as np

# Aqui estamos criando o valores de x e y
x = [num for num in range(1, 9)]
y = [round(np.log10(num) * 10) for num in range(1, 9)]
z = [randint(1, 9) for _ in range(8)]

# Configurando a linha
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abri', 'Maio', 'Junho', 'Julho', 'Agosto']

ax[0].bar(x, y, label='Números Aleatórios', color='plum', zorder=2)

ax[0].set_title('Gráfico de Exemplo 1', fontsize=18)
ax[0].set_xlabel('Meses', fontsize=14)
ax[0].set_xticks(x)
ax[0].set_xticklabels(meses)
plt.xticks(fontsize=14)
ax[0].set_ylabel('Quantidade', fontsize=14)

ax[0].grid(visible=True, axis='both')
ax[0].legend(fontsize=12)


ax[1].plot(x, z, label='Logaritmo de 1 à 9', color='sienna', linestyle='dashed', linewidth=1, marker='o')

ax[1].set_title('Gráfico de Exemplo 2', fontsize=18)
ax[1].set_xlabel('Meses', fontsize=14)
ax[1].set_xticks(x)
ax[1].set_xticklabels(meses)
plt.xticks(fontsize=14)
ax[1].set_ylabel('Quantidade', fontsize=14)

ax[1].grid(visible=True, axis='both')
ax[1].legend(loc='upper center',fontsize=12)


plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16, 6))
from numpy.random import randint
import numpy as np

# Aqui estamos criando o valores de x e y
x = [num for num in range(1, 9)]
y = [round(np.log10(num) * 10) for num in range(1, 9)]
z = [randint(1, 9) for _ in range(8)]

# Configurando a linha
ax.set_facecolor('black')
ax.plot(x, y, label='Números Aleatórios', color='g', linestyle='dashed', linewidth=1, marker='d')
ax.plot(x, z, label='Logaritmo de 1 à 9', color='sienna', linestyle='dashed', marker='8')

# Configurando a figura
ax.set_title('Gráfico de Exemplo', fontsize=22)
ax.set_xlabel('Meses', fontsize=18)
ax.set_xticks(x)
ax.set_xticklabels(
    ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto']
)
plt.xticks(fontsize=14)
ax.set_ylabel('Quantidade', fontsize=14)
plt.grid(visible=True, axis='y')
plt.legend()

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16, 6))
from numpy.random import randint

# Aqui estamos criando o valores de x e y
x = [num for num in range(1, 9)]
y = [randint(1, 9) for _ in range(8)]

# Configurando a linha
ax.plot(x, y, color='k', linestyle='dashed', linewidth=1, marker='d')

# Configurando a figura
ax.set_title('Gráfico de Exemplo', fontsize=18)
ax.set_xlabel('Meses', fontsize=18)
ax.set_xticks(x)
ax.set_xticklabels(
    ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago']
)
plt.xticks(fontsize=12)
ax.set_ylabel('Quantidade', fontsize=14)

plt.show()
```

```python
import pandas as pd

dados = {
    'X': [1,2,3,4,5,6],
    'Y': [120, 120, 130, 145, 118, 125],
    'Z': [95, 54, 86, 77, 90, 81]
}

base = pd.DataFrame(dados)
```

```python
base
```

```python
fig, ax = plt.subplots(figsize=(16, 6))

w=0.2
ax.bar(base.X+w/2, base.Y, label='Y', width=w)
ax.bar(base.X-w/2, base.Z, label='Z', width=w)

for i in range(6):
  ax.annotate(
      base.Z[i],
      (base.X[i]-w/2, base.Z[i]),
      ha='center',
      xytext=(0,5),
      textcoords='offset points'
      )
  ax.annotate(
      base.Y[i],
      (base.X[i]+w/2, base.Y[i]),
      ha='center',
      xytext=(0,5),
      textcoords='offset points'
  )


plt.legend()
plt.show()
```
