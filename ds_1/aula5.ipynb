```python
dataset = r'https://raw.githubusercontent.com/rafaelpuyau/infinity_school/main/ds/datasets/atividade_casa.csv'
```

```python
df = pd.read_csv(dataset)
```

```python
df['Valor'] > 100
```

```python
# Primeira Pergunta
maior_100 = df['Valor'] > 100
df.loc[maior_100].count()
```

```python
# Segunda pergunta
funcao1 = (df['Produto'] == 'Saia')
funcao2 = (df['Valor'] < 70)

saias = df.loc[(funcao1 & funcao2)]
saias['Quantidade'].sum()
```
