```python
import pandas as pd
```

```python
dataset = r'https://raw.githubusercontent.com/rafaelpuyau/infinity_school/main/ds/datasets/wines.csv'
df = pd.read_csv(dataset)
```

```python
df.sample(5)
```

# Questão 1

###Quanto custa o vinho mais caro

```python
vinho_mais_caro = df['price'].max()
print(f'O vinho mais caro custa: {vinho_mais_caro}')
```

#Questão 2
###Vinho mais barato

```python
vinho_mais_barato = df['price'].min()
print(f'O vinho mais barato custa: {vinho_mais_barato}')
```

#Questão 3
### Qual o vinho mais caro

```python
nome_vinho_mais_caro = df.loc[(df['label'] == 'Wine Q') & (df['price'] == vinho_mais_caro)]
print(f'O vinho mais caro é: {nome_vinho_mais_caro["label"].to_frame()}')
```

#Questão 4
### Vinho mais barato

```python
# Wine Z
nome_vinho_mais_barato = df.loc[(df['label'] == 'Wine Z') & (df['price'] == vinho_mais_barato)]
print(f"O vinho mais barato é: {nome_vinho_mais_barato['label'].to_frame()}")
```
