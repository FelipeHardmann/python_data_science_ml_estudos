```python
dfs = dfs.apply(lambda nota: np.random.uniform(low = 5.0, high = 10.0, size = 8), axis='columns')
dfs
```

```python
pd.options.display.float_format = '{:,.2f}'.format
dfs
```

```python
dfs['Média'] = dfs.mean(axis='columns')
dfs
```

```python
dfs['Situação'] = dfs['Média'].apply(lambda media: 'AP' if media >= 7.0 else 'RP')
dfs
```

```python
def calcular_media(media: float) -> str:
  '''DocString'''
  if media >= 7:
    return 'AP'
  return 'RP'
```

```python
dfs['Situação'] = dfs['Média'].apply(calcular_media)
dfs
```
