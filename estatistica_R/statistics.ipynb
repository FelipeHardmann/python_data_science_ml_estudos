```python
from statistics import (
    variance, 
    stdev
)

praca = [6, 10, 11, 15, 16, 19, 22, 25, 27, 27, 29, 40, 40, 40, 50, 60, 65, 67, 76, 77]

print(f'Praça: {round(mean(praca), 1)}')
print(f'Praça: Variância: {round(variance(praca),2)}')
print(f'Desvio Padrão Praça: {round(stdev(praca),2)}')

flamengo = [20, 21, 21, 22, 22, 23, 24, 24, 24, 25, 25, 25, 27, 28, 28, 29, 29, 29, 29, 30]
print(f'Flamengo: {round(mean(flamengo), 1)}')
print(f'Flamengo: Variância: {round(variance(flamengo),2)}')
print(f'Desvio Padrão Flamengo: {round(stdev(flamengo), 2)}')
```

```python
dolares = [4, 2, 4, 1, 4, 4, 4, 4, 1, 1]
reais = [20, 10, 20, 5, 20, 20, 20, 20, 5, 5]

cv = lambda x: stdev(x)/mean(x)

print(f'Dólar: {round(stdev(dolares), 2)}')
print(f'Real: {round(stdev(reais), 2)}')

print(f'Média Dólar: {round(mean(dolares), 2)}')
print(f'Média Reais: {round(mean(reais), 2)}')

print(f'Coeficiente de variação dólar: {cv(dolares):.4f}')
print(f'Coeficiente de variação reais: {cv(reais):.4f}')
```
